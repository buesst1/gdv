```python
import pandas as pd
from plotnine import *
from plotnine.data import diamonds
```

```python
diamonds
```

```python
(ggplot(diamonds, 
        aes(x='carat', y='price')) 
        + geom_point(alpha=0.05)
        + theme_bw()
        + ggtitle("Diamantenpreis in AbhÃ¤ngigkeit des Gewichts")
        + xlab("Gewicht in Karat")
        + ylab("Peis in $")
    )
```

```python
(ggplot(diamonds, 
        aes(x='carat', y='price', color="color")) 
        + geom_point(alpha=0.1)
        + stat_smooth(method="lowess")
        + theme_bw()
        + ggtitle("Diamantenpreis in AbhÃ¤ngigkeit des Gewichts und Farbe")
        + xlab("Gewicht in Karat")
        + ylab("Peis in $")
    )
```

```python
(ggplot(diamonds, 
        aes(x='carat', y='price')) 
        + facet_wrap(("color"))
        + geom_point(aes(color="color"), alpha=0.1)
        + stat_smooth(method="lowess")
        + theme_bw()
        + ggtitle("Diamantenpreis in AbhÃ¤ngigkeit des Gewichts und Farbe")
        + xlab("Gewicht in Karat")
        + ylab("Peis in $")
        + theme(figure_size=(16, 8))
    )
```
