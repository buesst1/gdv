```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = sns.load_dataset("iris")
```

```python
species_unique = iris['species'].unique()

for i, species in enumerate(species_unique):
    subset = iris[iris['species'] == species]
    sns.scatterplot(data=subset, x="petal_width", y="petal_length", label=species, s=200 if i==2 else None)

plt.title("Scatterplot der BlÃ¼tenblatt-breite vs LÃ¤nge von verschiedenen Blumen")
plt.xlabel("BlÃ¼tenblatt-breite")
plt.ylabel("BlÃ¼tenblatt-lÃ¤nge")
plt.legend()
plt.show()
```

```python
types = ["d", "o", "x"]

for i, species in enumerate(iris.species.unique()):
    subset = iris[iris['species'] == species]
    sns.scatterplot(data=subset, x="petal_width", y="petal_length", label=species, marker=types[i], color="blue")

plt.title("Scatterplot der BlÃ¼tenblatt-breite vs LÃ¤nge von verschiedenen Blumen")
plt.xlabel("BlÃ¼tenblatt-breite")
plt.ylabel("BlÃ¼tenblatt-lÃ¤nge")
plt.legend()
plt.show()
```

```python
palette = sns.color_palette('husl', n_colors=len(iris['species'].unique()))

for i, species in enumerate(iris.species.unique()):
    subset = iris[iris['species'] == species]
    
    subset = iris[iris['species'] == species]

    sns.scatterplot(data=subset, x="petal_width", y="petal_length", color=palette[i], label=species)
    
    sns.kdeplot(x=subset["petal_width"], y=subset["petal_length"], levels=5, color=palette[i], fill=True, alpha=0.3, cut=2)

plt.title("Scatterplot/KDE der BlÃ¼tenblatt-breite vs LÃ¤nge von verschiedenen Blumen")
plt.xlabel("BlÃ¼tenblatt-breite")
plt.ylabel("BlÃ¼tenblatt-lÃ¤nge")
plt.legend(title='Species')
plt.show()
```
