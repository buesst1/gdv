```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = list(pd.read_excel("../data/je-d-08.03.02.03.xlsx", [0], header=4, index_col=0).values())[0]

df = df.iloc[2:14]
df = df.transpose()
df = df.reset_index(names="Jahr")

print("dtypes")
display(df.dtypes)
```

```python
df
```

```python
sns.scatterplot(df, x="Jahr", y="Sonne")
sns.lineplot(df, x="Jahr", y="Sonne")

plt.xlabel("Jahr")
plt.ylabel("Sonnenenergie in TJ")
plt.title("Verlauf der verbrauchten Energie generiert durch die Sonne in der Schweiz")
plt.show()
```

```python
copy = df.copy()
diff = np.diff(copy.Sonne.values)
copy = copy.iloc[1:]
copy["Sonnenenergie_differenz"] = diff

plt.figure(figsize=(10, 5))
ax = sns.barplot(copy, x="Jahr", y="Sonnenenergie_differenz", color="blue")
for item in ax.get_xticklabels():
    item.set_rotation(45)

plt.xlabel("Jahr")
plt.ylabel("Differenz zum Vorjahr der Sonnenenergie in TJ")
plt.title("Differenz zum Vorjahr der verbrauchten Sonnenenergie in der Schweiz")
plt.show()
```

```python
copy = pd.melt(df, id_vars=["Jahr"], value_vars=["Holz", "Biogase", "Sonne"])

copy = copy.groupby([pd.cut(copy.Jahr, 4), "variable"]).sum().drop(columns=["Jahr"], inplace=False).reset_index()

def round_interval(interval):
    """Rundet die Grenzen des Intervalls."""
    if pd.isna(interval) or not isinstance(interval, pd.Interval):
        return interval

    # Runden der linken und rechten Grenze des Intervalls
    left = round(interval.left)
    right = round(interval.right)

    # Erstellen eines neuen gerundeten Intervalls
    return pd.Interval(left, right, closed=interval.closed)

# Anwenden der Funktion auf die 'intervals' Spalte
copy['Jahr'] = copy['Jahr'].apply(round_interval)
copy.columns = ["Jahr", "Art", "value"]

display(copy)

sns.barplot(copy, x="Jahr", y="value", hue="Art")

plt.xlabel("Jahresintervall")
plt.ylabel("Energiemenge in TJ")
plt.title("Verbrauchte Energie nach Art über bestimmten Zeitintervallen")
plt.show()
```
