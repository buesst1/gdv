```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Datenset

```python
taxis = sns.load_dataset("taxis")

#fÃ¼ge price performance Spalte hinzu
taxis["price_performance"] = taxis.distance / taxis.fare

taxis
```

# Scatterplot

```python
#erstelle scatterplot
sns.regplot(taxis, x="distance", y="fare", line_kws={"color": "C1"}, scatter_kws={'alpha':0.3})

# Beschriften der Achsen und Setzen des Titels
plt.xlabel('gefahrene Strecke in km')
plt.ylabel('Fahrpreis in CHF')
plt.title('Scatterplot von Strecke vs Fahrpreis')
plt.show()
```

```python
#erstelle scatterplot
sns.regplot(taxis[taxis.distance != 0], x="distance", y="price_performance", lowess=True, line_kws={"color": "C1"}, scatter_kws={'alpha':0.3})

# Beschriften der Achsen und Setzen des Titels
plt.xlabel('gefahrene Strecke in km')
plt.ylabel('Km pro CHF in km')
plt.title('Scatterplot von Strecke vs Preis-leistungs VerhÃ¤ltnis')
plt.show()
```

# Histogramm

```python
sns.histplot(taxis, x="distance", bins=50, stat="probability")
```

# Boxplot

```python
sns.boxplot(taxis, x="pickup_borough", y="distance")
```
